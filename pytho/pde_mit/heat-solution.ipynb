```python
def u(x,t):
    u= (3*math.sin(math.pi*x)*math.exp(-(math.pi**2)*t)-math.sin(3*math.pi*x)*math.exp(-9*(math.pi**2)*t))/2
    return u
```

```python
def ua(xa,t):
    uar=[]
    for x in xa:
        uar.append(u(x,t))
    return uar
```

```python
def heatsolTypeI(x,t,N,nn,Bn):
    ff=0
    for m in range(0,N):
        Bnm=Bn[m]
        nnm=nn[m]
        ff=ff + Bnm*math.sin(nnm*math.pi*x)*math.exp(-math.pi**2*nnm**2*t)
    return ff
```

```python
def heatsolTypeIa(xa,ta,N,nn,Bn):
    sola=[]
    for t in ta:
        solt=[]
        for x in xa:
            solt.append(heatsolTypeI(x,t,N,nn,Bn))
        sola.append(solt) 
    return sola
```
