```python
#function ff=f3(x)
    #ff = math.pow(math.sin(math.pi*x),2)*(x>=1&x<=2) - math.pow(math.sin(math.pi*x),2)*(x>=-2&x<=-1)
def f3(x):
    ff=[]
    for xi in x:
        if(xi>=1 and xi<=2):
            ff.append(math.pow(math.sin(math.pi*xi),2))
        elif(xi>=-2 and xi<=-1):
            ff.append(- math.pow(math.sin(math.pi*x),2))
        else:
            ff.append(0)
    return ff
```

```python
#function ff=f1(x)
#ff=(x+1).*(x>=-1 & x<=0) + (1-2*x).*(x>0 & x<=1/2);

def  f1(x):
    ff=[]
    for xi in x:
        if(xi>=-1 and xi<=0):
            ff.append(xi)
        elif(xi>0 and xi<=1/2):
            ff.append(1-2*xi)
        else:
            ff.append(0)
    return ff
```
